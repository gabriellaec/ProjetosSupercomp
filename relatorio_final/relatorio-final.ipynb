```python
def multi_df(dic):
    # https://www.geeksforgeeks.org/nested-dictionary-to-multiindex-dataframe/
    reformed_dict = {}
    for outerKey, innerDict in dic.items():
        for innerKey, values in innerDict.items():
            reformed_dict[(outerKey,
                           innerKey)] = values

    multiIndex_df = pd.DataFrame(reformed_dict)
    return multiIndex_df
```

```python
def exe_dict(executaveis):
    dic = {}
    for x in executaveis:
        a_file = open(f"cache/{x}.pkl", "rb")
        output = pickle.load(a_file)
        a_file.close()
        dic_new = output
        dic.update(dic_new)
    return dic
```

```python
def create_df(executaveis, initial, val_range, step):
    dic = exe_dict(executaveis)
    df = pd.DataFrame(dic)
    multiIndex_df = pd.DataFrame(multi_df(dic))

    valores=[initial]
    for i in range (1,val_range):
        valores.append(valores[i-1]+step)
    multiIndex_df["tamanho-entrada"] = valores
    multiIndex_df = multiIndex_df.set_index("tamanho-entrada")
    return valores, df, multiIndex_df
```

## Smith Waterman

A etapa inicial do projeto implementa a Heurística de Alinhamento Local de Smith-
Waterman, aplicada a duas sequências de DNA fornecidas como input, calculando o score
máximo de alinhamento entre elas. Em seguida, reconstrói o alinhamento local ótimo das
duas sequências.


### Funcionamento

A Heurística de Alinhamento Local de Smith-Waterman consiste em um algoritmo que
segue os seguintes passos:
1. Inicializa uma matriz
n x
m com as linhas e colunas na posição 0 (em ambos os
eixos) contendo o valor 0
2. Para cada 1≤i≤n e 1≤j≤m:
4. Calcula a diagonal de acordo com a seguinte fórmula:
H[i-1,j-1] + w(a[i],b[j])
onde:
 * w(a[i],b[j])=2 se houve match,
 * w(a[i],b[j])= -1 se houve mismatch e
 * w(a[i],b[j])= -1 se houve gap
 
    
5. Calcula a deleção por meio de H[i-1,j] - 1
6. Calcular a inserção por meio de H[i,j-1] - 1
7. Atribui à posição da matriz H[i,j] o máximo dentre os valores;
   * diagonal
   * deleção
   * inserção
   * 0 (de forma a nunca ser negativo)
8. Por fim, retorna o valor máximo da matriz


Durante a construção da matriz, o caminho seguido foi armazenado em um struct, de
forma que, partindo da posição do score máximo encontrado, foi possível retornar no
caminho, o que permitiu reconstruir as sequências com o melhor alinhamento possível.
Este caminho foi determinado com a seguinte correspondência:
   * deleção = salto de cima para baixo
   * inserção = salto da esquerda para a direita
   * diagonal


Desta forma, foi armazenada no struct a movimentação necessária nos eixos X e Y, de
acordo com a descrição acima, além do valor calculado para cada posição.

## Busca local

Esta etapa inicial do projeto implementa a estratégia de Busca Local com Aleatoriedade,
aplicada a duas sequências de DNA fornecidas como input, calculando o score máximo de
alinhamento entre elas, por meio da divisão em subsequências. Tal estratégia de divisão
em problemas menores tem o intuito de diminuir o tempo de processamento,
principalmente para sequências muito grandes

### Funcionamento

O algoritmo funciona por meio dos seguintes passos:
1. Lê duas sequências de DNA com tamanho especificado (a - tamanho n e b - tamanho
m), contidas em um arquivo de entrada
2. Para gerar a subsequência, devemos fornecer um ponto de início da sequência original
(j) e o tamanho que esta subsequência terá (k). Estes valores foram gerados
aleatoriamente, no intervalo:
                    k, 1<=k<=m
e
                     j, 0<=j<=m-k
Por exemplo, se o tamanho sorteado da sequência de entrada AGCACACA for 3 e j=1, teríamos
a seguinte subsequência:

3. Gerou-se um número positivo p, correspondente ao número de subsequências da
entrada a que serão geradas para serem comparadas com sb. Neste projeto, foi
determinado que o valor de p deve ser o dobro de k, um valor
suficientemente grande para que se obtenha diversidade nas sequências aleatórias
geradas (principalmente para entradas maiores). É esperado que quanto maior o valor
de p, maior a chance de se encontrar um score melhor, porém maior o tempo de
execução. De acordo com a necessidade, os limiares de p podem ser
ajustados.
4. Foi sorteado um número i, 0<=i<=n (correspondente ao ponto de início na sequência
a) e p sequências sa com o tamanho k calculado previamente foram geradas. Estas
sequências foram armazenadas em um vetor para futura comparação de alinhamento.
5. Percorrendo o vetor das subsequências sa armazenadas, foram comparadas as bases
nitrogenadas em cada posição com a posição correspondente da subsequência sb. Os
scores de cada par (sa,sb) foram calculados com os pesos wmat=2, wmis=-1 e wgap=-1
6. O score máximo do cálculo do passo anterior e as subsequências correspondentes a ele
foram armazenados e retornados pela função
7. Para melhorar o valor máximo de score encontrado, o algoritmo foi rodado múltiplas
vezes (1000). Tal valor escolhido arbitrariamente segue a mesma lógica de p, ou seja,
quanto maior, maior a chance de se encontrar um score melhor, porém maior o tempo
de execução.

## Busca Exaustiva

### Metodologia

Na Busca Exaustiva, uma solução ótima de um problema é encontrada, por meio da geração de todas as soluções possíveis.

Sendo uma entrada composta por duas sequências de DNA (A e B), o código implementado funciona por meio das seguintes etapas:

1. Geração de todas as subsequências de A e de B
2. Cálculo dos alinhamentos de cada par de subsequencias de A e de B (com os pesos wmat, wmis e wgap). Foram utilizadas as seguintes estratégias para o cálculo:

    2.1. Alinhamento pelo método de Smith Waterman

    2.2. Score simples, por meio de uma comparação direta entre as duas subsequências, sendo a maior truncada para que ambas possuam o mesmo tamanho
    
    2.3. Uma mistura das duas estratégias descritas acima, sendo o score simples utilizado no cálculo de quando as sequências possuem o mesmo tamanho e o método de Smith Waterman, caso contrário.


3. A saída do programa será composta do score de alinhamento, além das subsequências correspondentes alinhadas.

## Busca Exaustiva multicore - OpenMP

Para a implementação da estratégia de paralelização foi escolhido o projeto de busca exaustiva. Este foi um dos que obteu um desempenho mais lento. Além disso, é bastante paralelizável, já que existem muitas subsequências cujo score deve ser calculado, o que é feito de forma sem a existência de dependências entre elas. Ou seja, o loop de cálculo de scores era ingenuamente paralelizável. 

Desta forma, identifica-se uma situação em que a mesma operação de cálculo de scores (lenta) é executada para todos os elementos de um conjunto de dados (grande), o que é uma situação ideal par a a implementação de um paralelismo de dados.



Portanto, o uso de threads para calcular scores para diferentes sequências se apresenta como uma boa opção.

A estratégia escolhida foi a implementação de um parallel for.
Para isso, as subsequências geradas foram armazenadas em um único vetor

Em uma primeira implementação, este grande vetor continha um índice incremental e as duas subsequências correspondentes, pertencentes a cada uma das entradas.

Em seguida, este vetor era percorrido dentro de um Parallel for. Em um novo vetor, os valores dos resultados eram armazenados no índice correspondente, de modo a não ocorrerem conflitos entre as threads.
Por fim, este vetor preenchido era percorrido e o melhor resultado era salvo, juntamente com as subsequências correspondentes.



Visando a otimizar esta implementação, foi implementada a estratégia de utilizar o parallel for com reduction, de forma que o melhor resultado entre as threads (valor máximo de score encontrado) fosse devolvido. 

No entanto, ainda existia um problema. O vetor gerado que continha todas as subsequências ficava muito grande, o que causava um estouro na memória. 

Para contornar isso, a estratégia utilizada foi calcular o score de forma paralelizada por blocos. Ou seja, foi determinado um tamanho máximo que o vetor de subsequências poderia alcançar. Quando este valor era atingido, os scores eram calculados, divididos em threads. Em seguida, o vetor com as combinações era esvaziado e o processo continuava, partindo para o próximo bloco, até que o tamanho do vetor das subsequências acabasse.

-------

## Busca exaustiva manycore - GPU


Nesta etapa, foi utilizada uma estratégia conjunta, que combinava a implementação com OpenMP, acrescentando a utilzação da GPU.

Desta forma, manteve-se a etapa de gerar um grande vetor com todos os pares de subsequências possíveis e calcular o score de forma paralelizada, por meio do OpenMP. Porém, agora o cálculo do score ocorre dentro da GPU.



Vale ressaltar, contudo, que em termos de velocidade a CPU é superior. O que diferencia a GPU é sua capacidade de alta vazão de dados (throughput), otimizados para o processamento de forma paralela.

Além disso, no caso desta implementação, o endereçamento de memória é separado, de forma que os dados precisam ser copiadas para a GPU, o que leva tempo.

Desta maneira, para tornar a implementação mais rápida, é necessário enviar vários dados por vez para serem processados simultaneamente na GPU. Isso foi feito usando a estratégia de enviar as subsequências cujo score deveria ser calculado em lotes de tamanho definido e depois retornar apenas o melhor resultado de todo o lote.

Assim, dentro dos threads lançados pelo parallel for, o vetor foi cortado, de modo que cada thread lançasse um lote para a GPU. Após calculado o score, a função responsável pelo cálculo na GPU retornava apenas o melhor dentre todos os scores encontrados. Por fim, o reduction implementado nas threads identificava apenas o valor máximo.

Depois que as threads acabavam seu processamento, este valor máximo era comparado com um valor máximo global, de forma a salvar apenas o melhor resultado.

Em seguida, o vetor com as combinações de subsequências era esvaziado e o processo se repetia até que todas as subsequências tenham sido processadas.

No entanto, percebe-se que existem combinações de subsequências cujo valor atingido nunca poderá ser o máximo, como por exemplo, aquelas de tamanho muito pequeno. Levando isso em consideração, foi implementada também uma estratégia de Branch & Bound para vetores de entrada maiores, de forma que, após o primeiro bloco de scores ser processado e um primeiro máximo obtido, este valor era comparado com o máximo possível de cada sequência. Se não fosse identificada a possibilidade de melhoria, tais subsequências não eram incluídas no cálculo.

Mesmo com essa nova estratégia implementada, o desempenho não foi bom, sendo pior do que as outras implementadas anteriormente

Quando a implementação do paralelismo foi retirada, o programa rodou mais rápido.
Uma possível explicação é que a implementação com OpenMP gerava as subsequencias em blocos em um vetor e depois disso enviava cada um desses blocos para a GPU.
Além do processamento mais lento causado pela geração do vetor com todos os pares subsequencias e a geração de batches menores para o cálculo do score, a transferência da GPU ocorria por partes e por blocos menores.
Considerando que este processo de transferência é muito lento, é de se esperar que quando ele ocorre múltiplas vezes com menos dados de cada vez o programa demore mais de fato.

Além disso, o OpenMP pode causar um problema de escalonamento dos threads. 

Normalmente o OpenMP usa o escalonamento estático, em que ele distribui as iterações em blocos entre as threads. Neste sentido, podem existir threads que tenham que calcular o score para subsequencias muito maiores do que outras threads.

Desta maneira, as threads que processarem sequências de tamanhos maiores levarão muito mais tempo do que aquelas que processam sequências de tamanho inferior.

Isso se torna mais evidente conforme o tamanho das sequências de entrada cresce, fazendo com que o tamanho das subsequencias geradas diverja cada vez mais.

Na prática, as threads que rodam mais rápido podem ficar ociosas.

Desta forma, o paralelismo acaba sendo prejudicado.

Uma possibilidade para melhorar o desempenho do programa seria realizar o processamento inteiramente dentro da GPU, de forma a paralelizar os processos mais efetivamente.

Desta forma, este valor deve ser ajustado conforme as necessidades de uso, se deseja-se favorecer o desempenho ou o tempo de execução.
